```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("HR_comma_sep.csv")
df
```

```python
dum1=pd.get_dummies(df.salary)
dum1
```

```python
df=pd.concat([df,dum1],axis='columns')
df
```

```python
df=df.drop(["Department","salary","last_evaluation","promotion_last_5years","lef"],axis=1)
df
```

```python
df1=df.drop(['high','low','medium'],axis=1)

df1
```

```python
regh=linear_model.LogisticRegression()
regm=linear_model.LogisticRegression()
regl=linear_model.LogisticRegression()
```

```python
regh.fit(df1,df[['high']])
regm.fit(df1,df[['medium']])
regl.fit(df1,df[['low']])
```

```python
regh.predict(df1)
```

```python
regh.score(df1,df[['high']])
```

```python
regm.score(df1,df[['medium']])
```

```python
regl.score(df1,df[['low']])
```
